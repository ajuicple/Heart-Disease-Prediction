```python
X = heart_df.drop("target", axis=1)
y = heart_df["target"]
```

```python
np.random.seed(42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
models = {"Logistic Regression": LogisticRegression(),
          "KNN": KNeighborsClassifier(),
          "Random Forest": RandomForestClassifier()}

def fit_score(models, X_train, X_test, y_train, y_test):
  np.random.seed(42)
  score_models={}
  for name, model in models.items():
    model.fit(X_train, y_train)
    score_models[name]=model.score(X_test, y_test)

    return score_models
```

```python
score_models = fit_score(models, X_train, X_test, y_train, y_test)
print(score_models)
```

```python
# Logistic Regression
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
log_models = LogisticRegression(max_iter=1000)
log_models.fit(X_train, y_train)
y_pred = log_models.predict(X_test)
acc = accuracy_score(y_test, y_pred)
conf_m = confusion_matrix(y_test, y_pred)
class_r = classification_report(y_test, y_pred)

print(acc)
print(conf_m)
print(class_r)
```

```python
# decision tree, random forest, svm, knn
tree_m = DecisionTreeClassifier(random_state=42)
forest_m = RandomForestClassifier(random_state=42)
svc_m = SVC()
knn_m = KNeighborsClassifier()

def te_m(model, X_train, X_test, y_train, y_test):
  model.fit(X_train, y_train)
  y_pred = model.predict(X_test)
  acc = accuracy_score(y_test, y_pred)
  conf_m = confusion_matrix(y_test, y_pred)
  class_r = classification_report(y_test, y_pred)

  return acc, conf_m, class_r

at_acc, at_conf, at_class = te_m(tree_m, X_train, X_test, y_train, y_test)
bt_acc, bt_conf, bt_class = te_m(forest_m, X_train, X_test, y_train, y_test)
ct_acc, ct_conf, ct_class = te_m(svc_m, X_train, X_test, y_train, y_test)
dt_acc, dt_conf, dt_class = te_m(knn_m, X_train, X_test, y_train, y_test)
print(at_acc, bt_acc, ct_acc, dt_acc)
```
